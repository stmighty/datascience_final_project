### 2.2.0 version 0

```python
count = 0
nan_country = 0
nan_organization = 0
nan_keyword = 0
nan_date = 0
import os
directory = '../../data/project'


titles = []
abstracts = []
publication_dates = []
affiliationses = []
keywordses = []
 
for subdir in os.listdir(directory):
    f1 = os.path.join(directory, subdir)
    #print(f1)
    num = 0
    for filename in os.listdir(f1) :
        f2 = os.path.join(f1, filename)
        with open(f2, 'r') as file:
            data = json.load(file)
        # Extract information from the JSON structure
        title = data['abstracts-retrieval-response']['item']['bibrecord']['head']['citation-title']
        abstract = data['abstracts-retrieval-response']['item']['bibrecord']['head']['abstracts']
        try :
            publication_date = data['abstracts-retrieval-response']['item']['bibrecord']['head']['source']['publicationdate']['date-text']['$']
        except :
            try :
                publication_date = data['abstracts-retrieval-response']['item']['bibrecord']['head']['source']['publicationdate']['date-text']
                #print(publication_date)
            except :
                print("error")
                print(data['abstracts-retrieval-response']['item']['bibrecord']['head']['source']['publicationdate'])
                publication_date = None
                nan_date += 1
                
                
                
                
                
                
                
                

       
        if(isinstance(data['abstracts-retrieval-response']['item']['bibrecord']['head']['author-group'], list)) :
            og_affiliations = []
            for aff in data['abstracts-retrieval-response']['item']['bibrecord']['head']['author-group'] :
                #print(aff['affiliation'])
                try :
                    og_affiliations.append(aff['affiliation'])
                except :
                    count += 1
            

        else :
            og_affiliations = [data['abstracts-retrieval-response']['item']['bibrecord']['head']['author-group']['affiliation']]
        
        affiliations = {}
        #print(og_affiliations)
        for i in og_affiliations :
            try :
                country = i['country']
            except :
                country = None
                nan_country += 1
                
            try :
                organization_og = i['organization']
            except :
                organization_og = None
                nan_organization += 1
                
            if(isinstance(organization_og, list)) :
                try :
                    organization = [org['$'] for org in organization_og]
                    organization = [organization[-1]]
                except :
                    organization = [None]
                    nan_organization += 1
               
            else :
                #print(organization_og)
                try :
                    organization = [organization_og['$']]
                except :
                    nan_organization += 1
                    organization = [None]
                    
                    
            if(country not in affiliations) :
                affiliations[country] = organization
            else :
                if(organization[0] not in affiliations[country]) :
                    affiliations[country] += organization
                    
        keywords = []
        keywords_og = data['abstracts-retrieval-response']['authkeywords']
        try :
            list_of_keywords = keywords_og['author-keyword']
            keywords = []
            for i in list_of_keywords :
                keywords.append(i['$'])
    
        except :
            nan_keyword += 1

        
        

                
            
            
        
        
        
        
        
        
        
        
        #print("Title :", title)
        #print("abstract :", abstract)
        #print("Publication Date:", publication_date)
        #print("affiliations :", affiliations)
        #print()
        titles.append(title)
        abstracts.append(abstract)
        publication_dates.append(publication_date)
        affiliationses.append(affiliations)
        keywordses.append(keywords)
        num += 1
        #if(num>=5) :
            #break
        
print("count :", count)
print("nan_country :", nan_country)
print("nan_organization", nan_organization)
print("nan_keyword :", nan_keyword)
print("nan_date :", nan_date)

print("length of titles is :", len(titles))
print("length of abstracts is :", len(abstracts))
print("length of publication_dates is :", len(publication_dates))
print("length of affiliationses is :", len(affiliationses))
print("length of keywordses is :", len(keywordses))
    
```

```python
import pandas as pd
df = pd.DataFrame({
    'Title': titles,
    'Abstract': abstracts,
    'Publication Date': publication_dates,
    'Affiliations': affiliationses,
    'Keywords': keywordses
})

# Write the DataFrame to a CSV file
df.to_csv('../../data/output_data.csv', index=False)
```
