```python
# Extract the key from the dictionary
key = list(dict_data.keys())
value = list(dict_data.values())

print(key)
print(value)
```

## Handle many-to-many of Article Table and Country Table

```python
df_Aff.head(10)
```

```python
for articleID, affiliation in df_Aff.head().iterrows():
  print(affiliation.tolist())
```

```python
collection = [] # [ [ArticleID, CountryID, CountryName ], ]

for tempArticleID, affiliation in df_Aff.iterrows():
  articleID = affiliation.tolist()[0]
  dict_data = ast.literal_eval(affiliation.tolist()[1])
  key = list(dict_data.keys())
  for country in key:
    if country is not None:
      countryID = CountryTable[ CountryTable['CountryName'] == country ].iloc[0].tolist()[0]
      collection.append( [articleID, countryID, country])
    else:
      collection.append( [articleID, None, None])
```

```python
collection
```

# Create ArticleCountry Table
1. KeyID
2. ArticleID
3. CountryID
4. CountryName

```python
# Create a DataFrame from the list of lists
ArticleCountryTable = pd.DataFrame(collection, columns=['ArticleID', 'CountryID', 'CountryName'])

# Add an auto-incrementing KeyID column
# This starts from 1 and increments for each row
ArticleCountryTable.insert(0, 'KeyID', range(1, len(ArticleCountryTable) + 1))
```

```python
ArticleCountryTable.head(10)
```

```python
ArticleCountryDateTable = pd.merge(ArticleCountryTable, ArticleTable[['ArticleID', 'PublicationDate', 'PublicationYear']], on='ArticleID', how='inner')
```

```python
ArticleCountryDateTable.to_csv('../../data/extract_table/ArticleCountryDateTable.csv', index=False)
```

```python
df_Aff.head(10)
```

```python
OrgNames = []

for tempArticleID, affiliation in df_Aff.iterrows():
  dict_data = ast.literal_eval(affiliation.tolist()[1])
  value = list(dict_data.values())
  for orgs in value:
    for org in orgs:
      if org not in OrgNames:
        OrgNames.append(org)
```
