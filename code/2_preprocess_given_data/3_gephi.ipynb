```python
df_map = pd.read_csv('../../data/extract_table/CountryTable.csv')
df_map
dictionary = {}
for ind in df_map.index :
    key = df_map['CountryName'].iloc[ind]
    value = df_map['CountryID'].iloc[ind]
    dictionary[key] = value
print(dictionary)
    
    
    
```

```python
sources = []
targets = []
for ind in df.index :
    affiliation_str = df['Affiliations'].iloc[ind]
    affiliation_dict = ast.literal_eval(affiliation_str)
    keys = list(affiliation_dict.keys())
    for i in range(len(keys)) :
        for j in range(i+1, len(keys)) :
            if(keys[i] != None and keys[j] != None and keys[i] != keys[j]) :
                sources.append(dictionary[keys[i]])
                targets.append(dictionary[keys[j]])
                


            

            
    
        
    
```

```python
print(len(sources))
print(len(targets))
```

```python
ids = [str(i) for i in range(1,679725)] 
    
```

```python
df = pd.DataFrame({
    'id' : ids,
    'source': sources,
    'targets': targets
})

# Write the DataFrame to a CSV file
df.to_csv('../../data/extract_table/edge.csv', index=False)
```
